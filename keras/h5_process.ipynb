# Check weights

```python
import h5py
import numpy as np
```

```python
!h5dump -n model_best.h5
```

```python
filename = 'model_best.h5'
f = h5py.File(filename, 'r')
```

```python
w_1 = f['/dense_1/dense_1/kernel:0']
w_2 = f['/dense_2/dense_2/kernel:0']
w_3 = f['/dense_3/dense_3/kernel:0']
b_1 = f['/dense_1/dense_1/bias:0']
b_2 = f['/dense_2/dense_2/bias:0']
b_3 = f['/dense_3/dense_3/bias:0']

weights = [w_1, w_2, w_3]
biases = [b_1, b_2, b_3]

for i in range(3):
    print("w_{}:\t{}\nb_{}:\t{}\t".format(i, weights[i].shape, i, biases[i].shape))
    
print(np.array(weights[0]))
```

# save weights

```python
!mkdir models_best
```

```python
for i in range(1, 4):
    np.save("models_best/w_{}".format(i), np.array(weights[i-1]))
    np.save("models_best/b_{}".format(i), np.array(biases[i-1]))
```

```python
w_1_load = np.load("models_best/w_1.npy")
assert np.allclose(w_1, w_1_load)
print(w_1_load)
```
