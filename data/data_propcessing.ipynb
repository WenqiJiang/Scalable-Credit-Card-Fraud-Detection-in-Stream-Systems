# Data reorginization

We are going to use several reorganization methods: subsampling negative samples, expanding positive samples, etc.

## Load data

```python
import numpy as np
```

```python
X_train = np.load("./origin_data/X_train.npy")
X_val = np.load("./origin_data/X_val.npy")
X_test = np.load("./origin_data/X_test.npy")
y_train = np.load("./origin_data/y_train.npy")
y_val = np.load("./origin_data/y_val.npy")
y_test = np.load("./origin_data/y_test.npy")

print(X_train.shape)
print(X_val.shape)
print(X_test.shape)
print(y_train.shape)
print(y_val.shape)
print(y_test.shape)
```

## Subsampling

find all frauds -> subsample equal amount of non-frauds

### Training Set

```python
count = np.sum(y_train)
print(count)

pos_index = np.array([i for i in range(X_train.shape[0]) if y_train[i] == 1])
print(pos_index)

X_train_pos = np.array([X_train[i] for i in range(X_train.shape[0]) if y_train[i] == 1])
X_train_neg = np.array([X_train[i] for i in range(X_train.shape[0]) if y_train[i] == 0])
print(X_train_pos.shape)
print(X_train_neg.shape)
```

```python
neg_idx = np.random.randint(0, high=X_train_neg.shape[0], size = count)
neg_samples = X_train_neg[neg_idx]
print(neg_idx)
print(neg_samples.shape, '\n', neg_samples)
```

```python
processed_X_train = np.zeros((count*2, X_train.shape[1]))
processed_y_train = np.zeros((count*2), dtype=np.int32)
for i in range(count):
    processed_X_train[2 * i] = X_train_pos[i]
    processed_X_train[2 * i + 1] = X_train_neg[neg_idx[i]]
    processed_y_train[2 * i] = 1
    processed_y_train[2 * i + 1] = 0
    
print(processed_X_train.shape, '\n', processed_X_train)
print(processed_y_train.shape, processed_y_train.dtype, '\n', processed_y_train)
```

### Validation Set

```python
count = np.sum(y_val)
print('count', count)

pos_index = np.array([i for i in range(X_val.shape[0]) if y_val[i] == 1])
print('pos_index', pos_index)

X_val_pos = np.array([X_val[i] for i in range(X_val.shape[0]) if y_val[i] == 1])
X_val_neg = np.array([X_val[i] for i in range(X_val.shape[0]) if y_val[i] == 0])
print('X_val_pos.shape', X_val_pos.shape)
print('X_val_neg.shape', X_val_neg.shape)

neg_idx = np.random.randint(0, high=X_val_neg.shape[0], size = count)
neg_samples = X_val_neg[neg_idx]
print('neg_idx', neg_idx)
print('neg_samples', neg_samples.shape, '\n', neg_samples)

processed_X_val = np.zeros((count*2, X_val.shape[1]))
processed_y_val = np.zeros((count*2), dtype=np.int32)
for i in range(count):
    processed_X_val[2 * i] = X_val_pos[i]
    processed_X_val[2 * i + 1] = X_val_neg[neg_idx[i]]
    processed_y_val[2 * i] = 1
    processed_y_val[2 * i + 1] = 0
    
print('processed_X_val', processed_X_val.shape, '\n', processed_X_val)
print('processed_y_val', processed_y_val.shape, processed_y_val.dtype, '\n', processed_y_val)
```

### test

```python
count = np.sum(y_test)
print('count', count)

pos_index = np.array([i for i in range(X_test.shape[0]) if y_test[i] == 1])
print('pos_index', pos_index)

X_test_pos = np.array([X_test[i] for i in range(X_test.shape[0]) if y_test[i] == 1])
X_test_neg = np.array([X_test[i] for i in range(X_test.shape[0]) if y_test[i] == 0])
print('X_test_pos.shape', X_test_pos.shape)
print('X_test_neg.shape', X_test_neg.shape)

neg_idx = np.random.randint(0, high=X_test_neg.shape[0], size = count)
neg_samples = X_test_neg[neg_idx]
print('neg_idx', neg_idx)
print('neg_samples', neg_samples.shape, '\n', neg_samples)

processed_X_test = np.zeros((count*2, X_test.shape[1]))
processed_y_test = np.zeros((count*2), dtype=np.int32)
for i in range(count):
    processed_X_test[2 * i] = X_test_pos[i]
    processed_X_test[2 * i + 1] = X_test_neg[neg_idx[i]]
    processed_y_test[2 * i] = 1
    processed_y_test[2 * i + 1] = 0
    
print('processed_X_test', processed_X_test.shape, '\n', processed_X_test)
print('processed_y_test', processed_y_test.shape, processed_y_test.dtype, '\n', processed_y_test)
```

### save them

```python
np.save('./subsamp_data/processed_X_train', processed_X_train)
np.save('./subsamp_data/processed_y_train', processed_y_train)
np.save('./subsamp_data/processed_X_val', processed_X_val)
np.save('./subsamp_data/processed_y_val', processed_y_val)
np.save('./subsamp_data/processed_X_test', processed_X_test)
np.save('./subsamp_data/processed_y_test', processed_y_test)
```

# oversample 10 times

## training set

```python
count = np.sum(y_train)
print(count)

pos_index = np.array([i for i in range(X_train.shape[0]) if y_train[i] == 1])
print('pos_index', pos_index)

X_train_pos = np.array([X_train[i] for i in range(X_train.shape[0]) if y_train[i] == 1])
X_train_neg = np.array([X_train[i] for i in range(X_train.shape[0]) if y_train[i] == 0])
print(X_train_pos.shape)
print(X_train_neg.shape)


processed_X_train = np.zeros((count * 20, X_train.shape[1]))
processed_y_train = np.zeros((count * 20), dtype=np.int32)

for j in range(10):
    neg_idx = np.random.randint(0, high=X_train_neg.shape[0], size = count)
    neg_samples = X_train_neg[neg_idx]
    
    for i in range(count):
        processed_X_train[j * 2 * count + 2 * i] = X_train_pos[i]
        processed_X_train[j * 2 * count + 2 * i + 1] = X_train_neg[neg_idx[i]]
        processed_y_train[j * 2 * count + 2 * i] = 1
        processed_y_train[j * 2 * count + 2 * i + 1] = 0
    
print(processed_X_train.shape, '\n', processed_X_train)
print(processed_y_train.shape, processed_y_train.dtype, '\n', processed_y_train)
```
